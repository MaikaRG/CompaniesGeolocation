```python
from pymongo import MongoClient
```

```python
def connectCollection(database, collection):
    client = MongoClient()
    db = client[database]
    coll = db[collection]
    return db, coll
```

```python
dbOne, collOne = connectCollection('getlocation','oldOffices')
```

```python
dbOf, collOf = connectCollection('getlocation','listogeoindex')
```

```python
one = list(collOne.find())
one[0]
```

```python
def getLejos (longitud,latitud):
    myOfOld = collOne.find(
        {"location":
         {"$near":
          {"$geometry":
           {"type":"Point",
            "coordinates": [longitud,latitud]
           },
           "$maxDistance":2000,
          }
         }
        }
    )
    return myOfOld
```

```python
myOff=list(collOf.find())
myOff[0]['offices']['longitude']
```

```python
for i in range(len(myOff)):
    a = getLejos(float(myOff[i]['offices']['longitude']),float(myOff[i]['offices']['latitude']))
    print(len(list(a)))
```
