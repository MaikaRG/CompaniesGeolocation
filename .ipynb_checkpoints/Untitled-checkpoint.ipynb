```python
import requests


def getCoordsFor(address):
    url = "https://devru-latitude-longitude-find-v1.p.rapidapi.com/latlon.php"
    querystring = {"location":address}
    headers = {
    'x-rapidapi-host': "devru-latitude-longitude-find-v1.p.rapidapi.com",
    'x-rapidapi-key': "4cea3751d1mshc7408ee7611d52ep1e8d35jsn200afd716a3e"
    }
    res = requests.request('GET',url,headers=headers, params=querystring)
    if res.status_code == 200:
        print(res.url)
        data = res.json()
        return (data.get("lat",0), data.get("lon",0))
    else:
        raise ValueError(f"Api returned {res.status_code}!=200")
```

```python
import requests


def getCoordsFor(address):
    url = f"https://geocode.xyz/{address}?json=1"
    res = requests.get(url)
    if res.status_code == 200:
        print(res.url)
        data = res.json()
        return (data.get("latt",0), data.get("longt",0))
    else:
        raise ValueError(f"Api returned {res.status_code}!=200")
```

```python
getCoordsFor("San Jose, 1261 Lincoln Ave")
```
