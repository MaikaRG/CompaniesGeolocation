```python
import requests

res = requests.get("https://es.numbeo.com/calidad-de-vida/clasificaciones-por-pa%C3%ADs")
html = res.text
html
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')
```

```python
scrap = soup.select('tbody')
for e in scrap:
    a = e
#print(a)
listaPaises=[]
listaPuntuacion=[]
for e in a.select('.cityOrCountryInIndicesTable'):
    t=e.text
    listaPaises.append(t)
for e in a.select('td')[2::11]:
    b = e.text
    listaPuntuacion.append(b)
listaPaises
listaPuntuacion
r=list(zip(listaPaises,listaPuntuacion))
r
```

```python
import carto
from pymongo import MongoClient
```

```python
def connectCollection(database, collection):
    client = MongoClient()
    db = client[database]
    coll = db[collection]
    return db, coll
```

```python
db, coll = connectCollection('Opciones','misOpciones')
```

```python
a = list(coll.find())
len(a)
print(a[0]['offices']['country'])
```

```python
listaOfis=[]
for e in range(len(a)):
    listaOfis.append(a[e]['offices']['country'])
listaOfis
```
